```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise A:

#### Scatter Plots | Normally Distributed

```python
# Normal Distribution
mean = 0    
std = 1 
N = 1000 #length
colors = np.random.rand(N)
size =(10 * np.random.rand(N))**2

x = np.random.normal(mean,std,size=N)
y = np.random.normal(mean,std,size=N)

plt.scatter(x, y, s=size, c=colors, alpha=0.5, marker='*', label='scatter')

# Label x & y axis, and title
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot | Normal Distribution')
plt.show() # Plot
```

#### Scatter Plots | Uniformly Distributed

```python
# Uniform Distribution
N = 1000
colors = np.random.rand(N)
size =(10 * np.random.rand(N))**2

x = np.random.uniform(-1, 1, size=N)
y = np.random.uniform(-1, 1, size=N)

plt.scatter(x, y, s=size, c=colors, alpha=0.5, marker='^', label='scatter')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot | Uniform Distribution')
plt.show()
```

## Exercise B:

#### Histogram | Normally Distributed | Showing Density

```python
x = np.random.normal(size = N)

fig, axes = plt.subplots(nrows=2, ncols=2) # shows 4 plots in 1
ax0, ax1, ax2, ax3 = axes.flatten()

# bin set to 20
ax0.hist(x, bins=20, histtype='bar', density=True, color="blue", edgecolor="black")
ax0.set_title('20 bin | bar | Showing Density')

# 50 bin
ax1.hist(x, bins=50, histtype='bar', density=True, color="blue", edgecolor="black")
ax1.set_title('50 bin | bar | Showing Density')

# 100 bin
ax2.hist(x, bins=100, histtype='bar', density=True, color="blue", edgecolor="black")
ax2.set_title('100 bin | bar | Showing Density')

# 200 bin
ax3.hist(x, bins=200, histtype='step', density=True, edgecolor="black") # Used 'step' histtype to switch up histograms
ax3.set_title('200 bin | step | Showing Density')

fig.tight_layout() # Fixes layout so titles are not on top of each other
plt.show()
```

#### Histogram | Normally Distributed | No Density

```python
x = np.random.normal(size = N)

fig, axes = plt.subplots(nrows=2, ncols=2)
ax0, ax1, ax2, ax3 = axes.flatten()

ax0.hist(x, bins=20, histtype='bar', color="green", edgecolor="black")
ax0.set_title('20 bin | bar')

ax1.hist(x, bins=50, histtype='bar', color="green", edgecolor="black")
ax1.set_title('50 bin | bar')

ax2.hist(x, bins=100, histtype='bar', color="green", edgecolor="black")
ax2.set_title('100 bin | bar')

ax3.hist(x, bins=200, histtype='step', edgecolor="black")
ax3.set_title('200 bin | step')

fig.tight_layout()
plt.show()
```
